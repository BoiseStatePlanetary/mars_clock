```python
from datetime import datetime, time, timedelta

class MissionTime(object):
    """
    Time(s) of sol for a given mission and sol
    """
    
    def __init__(self, time_str, mission, mission_sol, time_format="%H:%M:%S.%f"):
        """
        time_str (str): time in format given by time_format
        mission (str): name of the mission
        mission_sol (int): mission sol
        time_format (str, optional): specifies format of time string -
        see `here <https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior>`_ for details
        
        """
        cur_datetime = datetime.strptime(time_str, time_format)
        self.time = time(hour=cur_datetime.hour, 
                         minute=cur_datetime.minute, 
                         second=cur_datetime.second,
                         microsecond=cur_datetime.microsecond)
        
        self.mission = mission
        self.mission_sol = mission_sol
        
        # Calculate the mission sol represented by the time
        self.exact_time = mission_landing_dates[mission]['landing_date'] + timedelta(days=mission_sol) +\
            timedelta(hours=self.time.hour, minutes=self.time.minute, seconds=self.time.second, 
                      microseconds=self.time.microsecond)
        
    def __repr__(self):
        rep = 'Time: %02g:%02g:%02g.%g, Mission: %s, Mission Sol: %i' %\
            (self.time.hour, self.time.minute, self.time.second,
             self.time.microsecond, self.mission, self.mission_sol)
        return rep
```
